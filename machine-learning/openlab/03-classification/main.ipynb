```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./Social_Network_Ads.csv')
```

```python
purchased = df[df['Purchased'] == 1]
not_purchased = df[df['Purchased'] == 0]
```

```python
# Supondo que 'data' seja seu DataFrame com as colunas Age, EstimatedSalary e Purchased
# Exemplo de como plotar um gráfico de dispersão entre Age e EstimatedSalary
plt.figure(figsize=(10, 6))
plt.scatter(purchased['Age'], purchased['EstimatedSalary'])
plt.scatter(not_purchased['Age'], not_purchased['EstimatedSalary'])
plt.title('Dispersão entre Age e EstimatedSalary')
plt.xlabel('Age')
plt.ylabel('EstimatedSalary')
plt.grid(True)
plt.show()
```

```python
# Exemplo de como plotar um gráfico de dispersão entre Age e Purchased
plt.figure(figsize=(10, 6))
plt.scatter(purchased['Age'], purchased['Purchased'], c=purchased['Purchased'], cmap='viridis')
plt.scatter(not_purchased['Age'], not_purchased['Purchased'], c=not_purchased['Purchased'], cmap='viridis')
plt.title('Dispersão entre Age e Purchased')
plt.xlabel('Age')
plt.ylabel('Purchased')
plt.grid(True)
plt.show()
```

```python
# Exemplo de como plotar um gráfico de dispersão entre EstimatedSalary e Purchased
plt.figure(figsize=(10, 6))
plt.scatter(purchased['EstimatedSalary'], purchased['Purchased'], c=purchased['Purchased'], cmap='viridis')
plt.scatter(not_purchased['EstimatedSalary'], not_purchased['Purchased'], c=not_purchased['Purchased'], cmap='viridis')
plt.title('Dispersão entre EstimatedSalary e Purchased')
plt.xlabel('EstimatedSalary')
plt.ylabel('Purchased')
plt.grid(True)
plt.show()
```
