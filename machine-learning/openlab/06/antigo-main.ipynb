```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carregando o conjunto de dados Iris
iris_data = load_iris()

# Convertendo para DataFrame do Pandas
iris_df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
iris_df['target'] = iris_data.target
```

```python
iris_df.head()
```

```python
plt.hist(iris_df['target'])
plt.show()
```

```python
iris_0 = iris_df[iris_df['target'] == 0]
iris_1 = iris_df[iris_df['target'] == 1]
iris_2 = iris_df[iris_df['target'] == 2]
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(iris_0['sepal length (cm)'], iris_0['sepal width (cm)'], label='iris-setosa')
plt.scatter(iris_1['sepal length (cm)'], iris_1['sepal width (cm)'], label='iris-versicolor')
plt.scatter(iris_2['sepal length (cm)'], iris_2['sepal width (cm)'], label='iris-verginica')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.title('Distribuição de dimensões do SEPAL: length x width')
plt.legend()
plt.show()
```

```python

plt.figure(figsize=(10, 6))
plt.scatter(iris_0['petal length (cm)'], iris_0['petal width (cm)'], label='iris-setosa')
plt.scatter(iris_1['petal length (cm)'], iris_1['petal width (cm)'], label='iris-versicolor')
plt.scatter(iris_2['petal length (cm)'], iris_2['petal width (cm)'], label='iris-verginica')
plt.xlabel('pepal length (cm)')
plt.ylabel('pepal width (cm)')
plt.title('Distribuição de dimensões do PETAL: length x width')
plt.legend()
plt.show()
```
