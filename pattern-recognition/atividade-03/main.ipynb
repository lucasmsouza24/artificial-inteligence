```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
imagem = cv2.imread('Selecao.jpeg')
```

```python
imagemGray = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
face_frontal = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
face_recon = face_frontal.detectMultiScale(imagemGray, scaleFactor=1.3, minNeighbors=3)
# também é possível passa rum tamanho mínimo para o tamanho dos retangulos
# minSize=(50, 50)
```

```python
for(x, y, w, h) in face_recon:
    cv2.rectangle(imagem, (x, y), (x+w, y+h), (0, 255, 0), 4)
```

```python
plt.imshow(cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB))
plt.title('Isso é seleção')
plt.axis('off')
plt.show()
```
