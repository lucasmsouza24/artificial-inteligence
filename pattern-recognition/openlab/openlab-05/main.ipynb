```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
imagem = cv2.imread('Gatos2.jpg')
```

```python
def recognize_cat(image, scaleFactor=1.057, minNeighbors=1):
    imagemGray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    face_frontal = cv2.CascadeClassifier('./haarcascade_frontalcatface.xml')
    face_recon = face_frontal.detectMultiScale(imagemGray, scaleFactor=scaleFactor, minNeighbors=minNeighbors)
    # também é possível passa rum tamanho mínimo para o tamanho dos retangulos
    # minSize=(50, 50)

    for(x, y, w, h) in face_recon:
        cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 4)

    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title('Gatinhos')
    plt.axis('off')
    plt.show()
```

```python
recognize_cat(imagem)
```

```python

sininho_path = './sininho.jpeg'
sininho_img = cv2.imread(sininho_path)
recognize_cat(sininho_img, scaleFactor=5.1, minNeighbors=1)
```
