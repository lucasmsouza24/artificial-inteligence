# Reconhecimento de Padrões - Openlab 01

- Converter a imagem para tons de cinza;
- Aplicar o efeito “Blur” para remover o ruído e auxiliar a identificação das bordas;
- “Binarizar” a imagem para que se tenha apenas pixels pretos e brancos;
- Utilizar o detector de bordas para identificar os objetos;

### Imports

```python
import pandas as pd
import cv2
import matplotlib.pyplot as plt
```

### declarando função de exibição da imagem

```python
def display_image(image, text='Imagem'):
    plt.imshow(image, cmap='gray')
    plt.title(text)
    plt.axis('off')
    plt.show()
```

### Leitura da imagem

```python
image_path = './Tomates.jpg'
image = cv2.imread(image_path)
```

### Imagem em tons de cinza

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
display_image(image_gray, 'Imagem tons de cinza')
```

### Aplicando redução de nitidez (blur)

```python
image_blur = cv2.GaussianBlur(image_gray, (11, 11), 0)
display_image(image_blur, 'Imagem com redução de nitidez')
```

### Binarizando imagem

```python
_, image_binary = cv2.threshold(image_blur, 127, 255, cv2.THRESH_BINARY)
display_image(image_binary, 'Imagem binarizada')
```

### Reconhecendo bordas pelo algoritimo Canny

```python
edges = cv2.Canny(image_binary, 100, 200)
display_image(edges)
```
