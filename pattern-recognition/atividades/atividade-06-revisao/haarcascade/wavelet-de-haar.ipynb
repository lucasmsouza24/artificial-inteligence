```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('people.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
smile = cv2.CascadeClassifier('haarcascade_smile.xml')
face_frontal = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
smile_recon = smile.detectMultiScale(img_gray, scaleFactor = 3.01, minNeighbors = 50)
face_recon = face_frontal.detectMultiScale(img_gray, scaleFactor = 2.0, minNeighbors = 1)
```

```python
print('Sorrisos Encontrados: ', len(smile_recon))
print('Rostos Encontrados: ', len(face_recon))
```

```python
for (x, y, w, h) in smile_recon:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 4)

for (x, y, w, h) in face_recon:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 4)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
